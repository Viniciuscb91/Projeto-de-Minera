## Carregar Datasets para o ambiente de trabalho
Os Datasets v√™m de um rep√≥sitoro online no github. Nesta parte carregamos os mesmos a partir de uma request para podermos utiliz√°-los livremente para an√°lise:

```python
import pandas as pd
import requests
from io import StringIO

csv_urls = [
    ('Advanced', 'https://raw.githubusercontent.com/Nivaldo-Neto-IF/dados-minera/6ebee9563704ade59527c7ca15c21614a2451a8c/Advanced.csv'),
    ('Games', 'https://raw.githubusercontent.com/Nivaldo-Neto-IF/dados-minera/6ebee9563704ade59527c7ca15c21614a2451a8c/Games.csv'),
    ('NBA_GAMES', 'https://raw.githubusercontent.com/Nivaldo-Neto-IF/dados-minera/6ebee9563704ade59527c7ca15c21614a2451a8c/NBA_GAMES.csv'),
    ('player Shooting','https://raw.githubusercontent.com/Nivaldo-Neto-IF/dados-minera/09c6ae11fbb1deeba370379925582b1c41d6d156/Player_Shooting.csv'),
    ('Player_Totals','https://raw.githubusercontent.com/Nivaldo-Neto-IF/dados-minera/09c6ae11fbb1deeba370379925582b1c41d6d156/Player_Totals.csv'),
    ('Seasons_Stats', 'https://raw.githubusercontent.com/Nivaldo-Neto-IF/dados-minera/6ebee9563704ade59527c7ca15c21614a2451a8c/Seasons_Stats.csv'),
    ('all_seasons', 'https://raw.githubusercontent.com/Nivaldo-Neto-IF/dados-minera/6ebee9563704ade59527c7ca15c21614a2451a8c/all_seasons.csv'),
    ('player_data', 'https://raw.githubusercontent.com/Nivaldo-Neto-IF/dados-minera/6ebee9563704ade59527c7ca15c21614a2451a8c/player_data.csv'),
    ('Players', 'https://raw.githubusercontent.com/Nivaldo-Neto-IF/dados-minera/6ebee9563704ade59527c7ca15c21614a2451a8c/Players.csv')
]

loaded_dfs = {}

for i, (name, url) in enumerate(csv_urls):
    df_name = f'df{i+1}'
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = StringIO(response.text)

        try:
            df = pd.read_csv(data)
        except pd.errors.ParserError:
            data.seek(0)
            df = pd.read_csv(data, delimiter=';')

        globals()[df_name] = df
        loaded_dfs[df_name] = df

        print(f"Dataset '{name}' loaded successfully as '{df_name}'. First 5 rows:")
        display(df.head())
        print("-" * 30)

    except requests.exceptions.RequestException as e:
        print(f"Error accessing the GitHub URL for '{name}': {e}")
    except Exception as e:
        print(f"Error reading the CSV data for '{name}': {e}")
```

## E por fim buscamos separar as vari√°veis em qualitativas e quantitativas.

.




```python
def classificar_variaveis(df):
    tipos = []

    for col in df.columns:
        dtype = df[col].dtype

        if pd.api.types.is_numeric_dtype(df[col]):
            tipo = 'Quantitativa'
        elif pd.api.types.is_bool_dtype(df[col]):
            tipo = 'Booleana (Qualitativa)'
        elif pd.api.types.is_datetime64_any_dtype(df[col]) or pd.api.types.is_timedelta64_dtype(df[col]):
            tipo = 'Temporal'
        else:
            tipo = 'Qualitativa'

        tipos.append((col, tipo, str(dtype)))

    resultado = pd.DataFrame(tipos, columns=['Vari√°vel', 'Tipo', 'dtype'])
    return resultado
```

```python
for df_name, name in loaded_dfs.items():
    print(f"\nVariable classification for {name}:\n")
    display(classificar_variaveis(globals()[df_name]))
    print("-" * 30)
```
